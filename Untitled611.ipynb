```python
# =======================
# 1️⃣ IMPORTS
# =======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shap
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score
from sklearn.utils import class_weight

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, BatchNormalization, Dropout, Bidirectional, LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2

# =======================
# 2️⃣ DATA LOADING & CLEANING
# =======================
data = pd.read_csv('Modified_SQL_Datasettt.csv', encoding='ISO-8859-1')
data.columns = data.columns.str.strip()
query_column = 'Query'
label_column = 'Label'

data[query_column] = data[query_column].astype(str).str.lower().str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

# Balance dataset
counts = data[label_column].value_counts()
min_count = counts.min()
data_balanced = pd.concat([
    data[data[label_column] == label].sample(min_count, random_state=42)
    for label in counts.index
]).sample(frac=1, random_state=42)

# =======================
# 3️⃣ TOKENIZATION & PADDING
# =======================
max_words = 10000
max_len = 100

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(data_balanced[query_column])
sequences = tokenizer.texts_to_sequences(data_balanced[query_column])
X = pad_sequences(sequences, maxlen=max_len)
y = data_balanced[label_column].values

# =======================
# 4️⃣ FUNCTION: BUILD, TRAIN, VALIDATE & EVALUATE MODEL WITH SHAP
# =======================
def build_train_evaluate(X_train, X_test, y_train, y_test, split_name):
    weights = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
    class_weights = {i: weights[i] for i in range(len(weights))}

    model = Sequential([
        Embedding(input_dim=max_words, output_dim=128, input_length=max_len),
        Conv1D(128, 5, activation='relu', kernel_regularizer=l2(0.001)),
        MaxPooling1D(2),
        BatchNormalization(),
        Dropout(0.3),
        Bidirectional(LSTM(64, return_sequences=True, dropout=0.3)),
        LSTM(64, dropout=0.3),
        Dense(64, activation='relu', kernel_regularizer=l2(0.001)),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
    early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=15,
        batch_size=32,
        class_weight=class_weights,
        callbacks=[early_stop],
        verbose=1
    )

    # Evaluation
    loss, accuracy, auc = model.evaluate(X_test, y_test, verbose=0)
    y_probs = model.predict(X_test, verbose=0).flatten()
    fpr, tpr, thresholds = roc_curve(y_test, y_probs)
    best_thresh = thresholds[np.argmax(tpr - fpr)]
    y_pred = (y_probs >= best_thresh).astype(int)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_probs)

    print(f"\n===== Results for {split_name} Split =====")
    print(f"Loss: {loss:.4f}, Accuracy: {accuracy:.4f}, AUC: {auc:.4f}")
    print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}, ROC AUC: {roc_auc:.4f}")

    # ROC Curve
    plt.figure(figsize=(6,4))
    plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.4f}')
    plt.scatter(fpr[np.argmax(tpr - fpr)], tpr[np.argmax(tpr - fpr)], color='red', label=f'Best Threshold = {best_thresh:.2f}')
    plt.plot([0,1], [0,1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {split_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'SQLi'], yticklabels=['Normal', 'SQLi'])
    plt.title(f'Confusion Matrix - {split_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # SHAP Explainability
    print(f"\nRunning SHAP on {split_name} Split")
    shap.initjs()
    explainer = shap.Explainer(model, X_train[:100])
    shap_values = explainer(X_test[:10])
    shap.summary_plot(shap_values, X_test[:10], feature_names=[f'pos_{i}' for i in range(max_len)])
    shap.plots.waterfall(shap_values[0], max_display=15)
    shap.plots.bar(shap_values, max_display=15)

    return model

# =======================
# 5️⃣ RUN FOR MULTIPLE SPLITS
# =======================
splits = {
    '50/50': 0.5,
    '60/40': 0.6,
    '70/30': 0.7,
    '80/20': 0.8,
    '90/10': 0.9
}

for split_name, train_ratio in splits.items():
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=(1 - train_ratio), stratify=y, random_state=42)
    model = build_train_evaluate(X_train, X_test, y_train, y_test, split_name)

print("\n✅ Completed systematic CNN + BiLSTM SQL Injection Detection with Explainable AI using SHAP for all splits.")
```
